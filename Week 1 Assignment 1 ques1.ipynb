```python
import numpy as np
```

```python
from numpy import genfromtxt

my_data = genfromtxt('Starfish_Data.csv', delimiter=',')
```

```python
print (my_data)
```

```python
a=my_data[:,0]
b=my_data[:,1]
c=my_data[:,2]
d=(a + (2.5*b) + 0.17)/5 #log of distance calculated
print(d)

```

```python
distance=np.power(10,d)
distance
```

```python
#calculation of weighted mean
wt=np.dot(distance,c)
s=np.sum(c)
final=wt/s
final
```
