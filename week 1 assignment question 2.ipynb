```python
for i in range(len(hmxb.ra)):
    if hmxb.ra.iloc[i]>180:
        hmxb.ra.iloc[i]=hmxb.ra.iloc[i]-360
df3 = pd.DataFrame(hmxb.ra)
arr2 = df3.to_numpy()
arr2=np.radians(arr2)

df4 = pd.DataFrame(hmxb.dec)
arr4 = df4.to_numpy()
arr4=np.radians(arr4)
```

```python
obs=data[data.Astrosat_Flag == 1]
nobs=data[data.Astrosat_Flag == 0]
```

```python
for i in range(len(obs.ra)):
    if obs.ra.iloc[i]>180:
        obs.ra.iloc[i]=obs.ra.iloc[i]-360
df4= pd.DataFrame(obs.ra)
nike = df4.to_numpy()
nike=np.radians(nike)


df5= pd.DataFrame(obs.dec)
nike1 = df5.to_numpy()
nike1=np.radians(nike1)

```

```python
for i in range(len(nobs.ra)):
    if nobs.ra.iloc[i]>180:
        nobs.ra.iloc[i]=nobs.ra.iloc[i]-360
df6= pd.DataFrame(nobs.ra)
aro = df6.to_numpy()
aro=np.radians(aro)

df7= pd.DataFrame(nobs.dec)
aro1 = df7.to_numpy()
aro1=np.radians(aro1)
aro
aro1
```
